```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
businesses = pd.read_json(r'c:\Users\kasam\OneDrive\Desktop\yelp_dataset_challenge_academic_dataset/yelp_academic_dataset_business.json', lines=True, orient='columns')
```

```python
print("The business details dataset")
businesses.head(5)
```

```python
# Ratings,reviews and locations 
company_rating=businesses.groupby(['name']).stars.mean()
company_review_count=businesses.groupby(['name']).review_count.sum()
company_city_count=businesses.groupby(['name']).city.nunique()
result = pd.concat([company_review_count, company_rating,company_city_count], axis=1, join='inner')
result.sort_values(by=['city'], ascending=False).head(10)
```

```python
# Selecting mcdonalds for review analysis
mcdonalds=businesses.loc[businesses['name'] == "McDonald's"]
mcdonalds[['name','business_id','review_count','stars']].sort_values(['review_count','stars'], ascending=[False,False])
```

```python
#For Tacobell
Taco=businesses.loc[businesses['name'] == "Taco Bell"]
Taco[['name','business_id','review_count','stars']].sort_values(['review_count','stars'], ascending=[False,False])
```

```python
#For Subway
Subway=businesses.loc[businesses['name'] == "Subway"]
Subway[['name','business_id','review_count','stars']].sort_values(['review_count','stars'], ascending=[False,False])
```
