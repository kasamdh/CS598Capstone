```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
businesses = pd.read_json(r'c:\Users\kasam\OneDrive\Desktop\yelp_dataset_challenge_academic_dataset/yelp_academic_dataset_business.json', lines=True, orient='columns')
```

```python
print("The business details dataset")
businesses.head(5)
```

```python
# Ratings,reviews and locations 
company_rating=businesses.groupby(['name']).stars.mean()
company_review_count=businesses.groupby(['name']).review_count.sum()
company_city_count=businesses.groupby(['name']).city.nunique()
result = pd.concat([company_review_count, company_rating,company_city_count], axis=1, join='inner')
result.sort_values(by=['city'], ascending=False).head(10)
```

```python
# Selecting mcdonalds for review analysis
mcdonalds=businesses.loc[businesses['name'] == "McDonald's"]
mcdonalds[['name','business_id','review_count','stars']].sort_values(['review_count','stars'], ascending=[False,False])
```

```python
#For Tacobell
Taco=businesses.loc[businesses['name'] == "Taco Bell"]
Taco[['name','business_id','review_count','stars']].sort_values(['review_count','stars'], ascending=[False,False])
```

```python
#For Subway
Subway=businesses.loc[businesses['name'] == "Subway"]
Subway[['name','business_id','review_count','stars']].sort_values(['review_count','stars'], ascending=[False,False])
```

```python
# Data for McDonald's
mcdonalds = businesses.loc[businesses['name'] == "McDonald's"]
mcdonalds_review_count = mcdonalds['review_count'].values
mcdonalds_stars = mcdonalds['stars'].values

# Data for Taco Bell
taco_bell = businesses.loc[businesses['name'] == "Taco Bell"]
taco_bell_review_count = taco_bell['review_count'].values
taco_bell_stars = taco_bell['stars'].values

# Data for Subway
subway = businesses.loc[businesses['name'] == "Subway"]
subway_review_count = subway['review_count'].values
subway_stars = subway['stars'].values

# Plotting
fig, axs = plt.subplots(3, 2, figsize=(12, 12))

# McDonald's
axs[0, 0].bar(np.arange(len(mcdonalds_review_count)), mcdonalds_review_count, color='blue')
axs[0, 0].set_title("McDonald's Review Count")
axs[0, 0].set_xlabel('Locations')
axs[0, 0].set_ylabel('Review Count')

axs[0, 1].bar(np.arange(len(mcdonalds_stars)), mcdonalds_stars, color='green')
axs[0, 1].set_title("McDonald's Average Rating")
axs[0, 1].set_xlabel('Locations')
axs[0, 1].set_ylabel('Stars')

# Taco Bell
axs[1, 0].bar(np.arange(len(taco_bell_review_count)), taco_bell_review_count, color='blue')
axs[1, 0].set_title("Taco Bell Review Count")
axs[1, 0].set_xlabel('Locations')
axs[1, 0].set_ylabel('Review Count')

axs[1, 1].bar(np.arange(len(taco_bell_stars)), taco_bell_stars, color='green')
axs[1, 1].set_title("Taco Bell Average Rating")
axs[1, 1].set_xlabel('Locations')
axs[1, 1].set_ylabel('Stars')

# Subway
axs[2, 0].bar(np.arange(len(subway_review_count)), subway_review_count, color='blue')
axs[2, 0].set_title("Subway Review Count")
axs[2, 0].set_xlabel('Locations')
axs[2, 0].set_ylabel('Review Count')

axs[2, 1].bar(np.arange(len(subway_stars)), subway_stars, color='green')
axs[2, 1].set_title("Subway Average Rating")
axs[2, 1].set_xlabel('Locations')
axs[2, 1].set_ylabel('Stars')

plt.tight_layout()
plt.show()
```

```python
# Data for McDonald's
mcdonalds_positive_reviews = mcdonalds[mcdonalds['stars'] >= 3]
mcdonalds_negative_reviews = mcdonalds[mcdonalds['stars'] < 3]

# Data for Taco Bell
taco_bell_positive_reviews = taco_bell[taco_bell['stars'] >= 3]
taco_bell_negative_reviews = taco_bell[taco_bell['stars'] < 3]

# Data for Subway
subway_positive_reviews = subway[subway['stars'] >= 3]
subway_negative_reviews = subway[subway['stars'] < 3]

# Plotting
fig, axs = plt.subplots(3, 2, figsize=(12, 12))

# McDonald's
axs[0, 0].hist(mcdonalds_positive_reviews['stars'], bins=[1, 2, 3, 4, 5], color='green', alpha=0.7)
axs[0, 0].set_title("McDonald's Positive Reviews Distribution")
axs[0, 0].set_xlabel('Star Rating')
axs[0, 0].set_ylabel('Frequency')

axs[0, 1].hist(mcdonalds_negative_reviews['stars'], bins=[1, 2, 3, 4, 5], color='red', alpha=0.7)
axs[0, 1].set_title("McDonald's Negative Reviews Distribution")
axs[0, 1].set_xlabel('Star Rating')
axs[0, 1].set_ylabel('Frequency')

# Taco Bell
axs[1, 0].hist(taco_bell_positive_reviews['stars'], bins=[1, 2, 3, 4, 5], color='green', alpha=0.7)
axs[1, 0].set_title("Taco Bell Positive Reviews Distribution")
axs[1, 0].set_xlabel('Star Rating')
axs[1, 0].set_ylabel('Frequency')

axs[1, 1].hist(taco_bell_negative_reviews['stars'], bins=[1, 2, 3, 4, 5], color='red', alpha=0.7)
axs[1, 1].set_title("Taco Bell Negative Reviews Distribution")
axs[1, 1].set_xlabel('Star Rating')
axs[1, 1].set_ylabel('Frequency')

# Subway
axs[2, 0].hist(subway_positive_reviews['stars'], bins=[1, 2, 3, 4, 5], color='green', alpha=0.7)
axs[2, 0].set_title("Subway Positive Reviews Distribution")
axs[2, 0].set_xlabel('Star Rating')
axs[2, 0].set_ylabel('Frequency')

axs[2, 1].hist(subway_negative_reviews['stars'], bins=[1, 2, 3, 4, 5], color='red', alpha=0.7)
axs[2, 1].set_title("Subway Negative Reviews Distribution")
axs[2, 1].set_xlabel('Star Rating')
axs[2, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
